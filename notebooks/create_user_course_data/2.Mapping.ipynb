```python
import pandas as pd
import csv
```

<h1>1. Create Mapping class ü•¥</h1>

```python
class mapping:
    def __init__(self, data_path): # Constructor
        self.data_path = data_path

    def load_data(self): # Data loader
        self.df = pd.read_csv(self.data_path)

    def mapping(self, col):
        map = {}

        int_col = pd.DataFrame(self.df[col].unique()).rename(columns = {0: col}) # L·∫•y c√°c gi√° tr·ªã unique c·ªßa c·ªôt

        int_col[col] = int_col[col].apply(lambda x: x.strip(' ')) # Format c·ªôt (cho ch·∫Øc ch·∫Øn)

        for i in range(len(int_col[col])):
            map[int_col[col][i]] = i # T·∫°o mapping

        print(map)

        return map

    def create_mapping(self):
        self.course_map = self.mapping('course') # T·∫°p map cho c·ªôt course
        self.user_map = self.mapping('user') # T·∫°p map cho c·ªôt user

    def apply_mapping(self):
        self.df['course'] = self.df['course'].apply(lambda x: self.course_map[x]) # Chuy·ªÉn ƒë·ªïi c·ªôt user
        self.df['user'] = self.df['user'].apply(lambda x: self.user_map[x]) # Chuy·ªÉn ƒë·ªïi c·ªôt course

    def map_to_csv(self, user_path, course_path):
        with open(user_path, 'w') as csv_file: # Ghi user map v√†o ƒë∆∞·ªùng d·∫´n user_path
            writer = csv.writer(csv_file)

            for key, value in self.user_map.items():
                writer.writerow([key, value])

            csv_file.close()

        with open(course_path, 'w') as csv_file: # Ghi course map v√†o ƒë∆∞·ªùng d·∫´n course_path
            writer = csv.writer(csv_file)

            for key, value in self.course_map.items():
                writer.writerow([key, value])

            csv_file.close()

    def data_to_csv(self, data_path):
        self.df.to_csv(data_path, index = False) # L∆∞u d·ªØ li·ªáu
```

<h1>2. Using class üò≥</h1>

```python
data_path = '/content/drive/MyDrive/Nhom12/ƒêAMH/Project/Data/transformed_data/n_core_user_course.csv'
user_map = '/content/drive/MyDrive/Nhom12/ƒêAMH/Project/Data/transformed_data/map/user_map.csv'
course_map = '/content/drive/MyDrive/Nhom12/ƒêAMH/Project/Data/transformed_data/map/course_map.csv'

map_writer = mapping(data_path)
map_writer.load_data()
map_writer.create_mapping()
map_writer.apply_mapping()
map_writer.map_to_csv(user_map, course_map)
map_writer.data_to_csv('/content/drive/MyDrive/Nhom12/ƒêAMH/Project/Data/transformed_data/mapped/mapped_n_core_user_course.csv')
```

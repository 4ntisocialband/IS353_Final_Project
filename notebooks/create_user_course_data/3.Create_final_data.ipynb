<h1>1. Create Mapping class ü•¥</h1>

```python
class create_df:
    def __init__(self, data_path): # Constructor
        self.data_path = data_path

    def create_df(self): # Data loader
        df = pd.read_csv(self.data_path)

        df['enroll_time'] = pd.to_datetime(df['enroll_time'])

        df = df.sort_values(by=['user', 'enroll_time']).groupby('user').agg({
            'course': lambda x: list(x),
            'enroll_time': lambda x: list(x)
        }).reset_index()

        def get_test_label(x): # L·∫•y kh√≥a h·ªçc g·∫ßn nh·∫•t user h·ªçc l√†m test_label
            return x[-1]

        def get_val_label(x): # L·∫•y kh√≥a h·ªçc g·∫ßn nh√¨ user h·ªçc l√†m val label
            return x[-2]

        def get_feature(x): # L·∫•y c√°c kh√≥a h·ªçc c√≤n l·∫°i l√†m feature cho user
            return x[:-2]

        df['feature'] = df['course'].apply(get_feature)
        df['val_label'] = df['course'].apply(get_val_label)
        df['test_label'] = df['course'].apply(get_test_label)

        df['feature_time'] = df['enroll_time'].apply(get_feature)
        df['val_time'] = df['enroll_time'].apply(get_val_label)
        df['test_time'] = df['enroll_time'].apply(get_test_label)
        df = df.reset_index()

        self.df = df

    def get_df(self):
        return self.df

    def to_csv(self, train_path, val_path, test_path):
        # L∆∞u t·∫≠p train
        train_df = self.df[['user', 'feature', 'feature_time']]
        train_df.to_csv(train_path, index=False)
        # L∆∞u t·∫≠p validation
        val_df = self.df[['user', 'val_label', 'val_time']]
        val_df.to_csv(val_path, index=False)
        # L∆∞u t·∫≠p test
        test_df = self.df[['user', 'test_label', 'test_time']]
        test_df.to_csv(test_path, index=False)
```

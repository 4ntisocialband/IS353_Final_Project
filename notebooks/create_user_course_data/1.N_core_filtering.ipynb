<h1>1. Create n_core_filter class ü•¥</h1>

```python
class n_core_filter:
    def __init__(self, user_path, n_core): # Constructor
        self.user_path = user_path
        self.n_core = n_core

    def load_data(self): # Load d·ªØ li·ªáu
        df = pd.read_csv(self.user_path,
                         encoding = 'utf-8')

        self.df = df.reset_index(drop = True)

    def sampling_user(self, nums):
        sampling_user = self.df['user'].unique().tolist()
        np.random.shuffle(sampling_user)
        sampling_user = sampling_user[:nums]
        self.df = self.df[self.df['user'].isin(sampling_user)]

    def filtering(self):
        pre_len = -1 # Bi·∫øn l∆∞u tr·ªØ ƒë·ªô d√†i tr∆∞·ªõc khi l·ªçc n_core c·ªßa df, ƒë∆∞·ª£c kh·ªüi t·∫°o -1 ƒë·ªÉ b·∫Øt ƒë·∫ßu v√≤ng l·∫∑p

        while pre_len != self.df.shape[0]: # Khi pre_len tr√πng v·ªõi ƒë·ªô d√†i df (t·ª©c l√† kh√¥ng c√≤n user, course c√≥ th·ªÉ l·ªçc) k·∫øt th√∫c v√≤ng l·∫∑p:
            pre_len = self.df.shape[0]  # L∆∞u ƒë·ªô d√†i tr∆∞·ªõc khi l·ªçc

            self.df = self.df.groupby('user').filter(lambda x: len(x) >= self.n_core) # L·ªçc user c√≥ s·ªë course >= n_core
            self.df = self.df.groupby('course').filter(lambda x: len(x) >= self.n_core) # L·ªçc course c√≥ s·ªë user >= n_core

    def get_df(self):
        return self.df

    def to_csv(self, path):
        self.df.to_csv(path, index = False)
```

<h1>2. Using class üò≥</h1>

```python
data_path = '/content/drive/MyDrive/Nhom12/ƒêAMH/Project/Data/transformed_data/user_course.csv'

df = n_core_filter(data_path, 5)
df.load_data()
df.filtering()

to_see = df.get_df()
to_see.head()
```
